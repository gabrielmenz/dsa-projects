### 4- Como foi a evolução de incidentes por mês ao longo do tempo? ###

```python
df.head()
```

```python
df['month_of_incident'] = df['date_of_incident'].dt.month

monthly_totals = df.groupby(['month_of_incident', 'incident_event_type']).size().reset_index(name='monthly_total')
print(monthly_totals)

monthly_totals_summary = monthly_totals.groupby('month_of_incident')['monthly_total'].sum().reset_index(name='total_incidents')
print(monthly_totals_summary)

monthly_totals_summary['month_of_incident'] = pd.to_datetime(monthly_totals_summary['month_of_incident'], format='%m').dt.strftime('%b')
monthly_totals_summary
```

```python
ax = monthly_totals_summary.plot(kind='bar',
                                 x='month_of_incident',
                                 y='total_incidents',
                                 color='orange',
                                 legend=False)


plt.title('Total Incidents per Month')
plt.xlabel('Month')
plt.xticks(rotation=60)
plt.ylabel('Total Incidents')


for bar in ax.patches:
    bar_height = bar.get_height()
    label_position = (bar.get_x() + bar.get_width() / 2., bar_height)
    ax.annotate(str(bar_height),
                xy=label_position,
                ha='center',
                va='center',
                xytext=(0, 8),
                textcoords='offset points')


plt.show()
```
